```python
bottles = 99
for i in range(bottles, 0, -1):
    if i > 2:
        print("%i bottles of beer on the wall, %i bottles of beer!" %(bottles, bottles))
        bottles -= 1
        print("If one of those bottles should happen to fall, %i bottles of beer on the wall!\n" % bottles)
    elif i == 2:
        print("%i bottles of beer on the wall, %i bottles of beer!" %(bottles, bottles))
        bottles -= 1
        print("If one of those bottles should happen to fall, %i bottle of beer on the wall!\n" % bottles)
    elif bottles == 1:
        print("1 bottle of beer on the wall, 1 bottle of beer!")
        print("If that one bottle should happen to fall, what a waste of alcohol!.")
    
```
